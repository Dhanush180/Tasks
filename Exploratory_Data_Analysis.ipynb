### Sales and Profits by Region

```python
#Sales by Region
s_reg = df.groupby('Region')['Sales'].sum().sort_values(ascending= False)
s_reg.plot(kind='pie')
```

```python
#Profit by Region
prof_by_reg = df.groupby('Region')['Profit'].sum().sort_values(ascending= False)
prof_by_reg.plot(kind='pie')
```

### Sales by state

```python
#Sales by State
sales = df.groupby('State')['Sales'].sum().sort_values(ascending= False)
#top 10 states by sales 
sales[:10].plot(kind= 'bar')
```

```python
# statewise graph plot
state=df.groupby(['State'])['Sales'].sum()
state.sort_values(ascending=False,inplace=True)
fig, ax=plt.subplots(figsize=(20,10))
plt.title('state with Sales',fontsize=35)

state.plot.bar()
```

### Profits by State

```python
profits = df.groupby('State')['Profit'].sum().sort_values(ascending= False)

#top 10 states by profit
profits[:10].plot(kind = 'bar')
plt.ylabel('Profit')
```

```python
#bottom 10 states by profit
profits[39:].plot(kind = 'bar')
plt.ylabel('Profit')
```

```python
# Calculation information of each states of Profit
df.groupby("State").Profit.agg(["sum","mean","min","max","count","median","std","var"])
```

### Sales Profit per Category

```python
#statewise graph plot
sales_profit=df.groupby(['Sub-Category'])['Profit','Sales'].agg(['sum'])
sales_profit.plot.bar(rot=0,figsize=(20,10))

plt.title('Sales Profit per Category',fontsize=35)
plt.show(block=True)
```

### Sub category vs Region Analysis

```python
plt.figure(figsize =(30,10))
sns.countplot(x='Sub-Category', hue='Region', data=df)
plt.xticks(rotation= 45, fontsize= 18)
plt.yticks(fontsize= 18)
plt.xlabel('Sub-Category',fontsize= 25)
plt.ylabel('Count',fontsize= 25)
plt.legend(fontsize='20')
```

```python
df2= df.loc[:,['Sales', 'Quantity','Discount','Profit']]
sns.pairplot(df2)
```

From all the above analysis we can conclude that

1.Profits and Sales are positively correlated in all the regions, i.e with increase in sales, Profits also increase.

2.No significant correlation between Quantity and Profit.

3.Furniture category is bearing significant losses in Central region and small losses in South & East region.

4.When Discount offered is less Profits rise and when the Discount increases, losses are suffered.

5.Whereas Pennsylvania, Ohio, Texas are among the States bearing highest losses.

6.Binders and Papers have the highest sales across all the Regions, whereas Macines and Copiers have least sales

7.State with the highest amount of sales is California followed by New York, Texas, and so on

8.Profit margin in California and New York is very high compared to other states.

To increase profits there should be a balance between the sales and the discounts offered.

Profits in furniture segment can be incresed by decresing the dicounts on them.

Sales in the South and Central Region should be focused.

States with lesser sales(West Virginia, Dakota, Wyoming etc.) should be focused.
